```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Week 3 Kifiya/MachineLearningRating_v3.txt', delimiter = "|")

# Check the structure of the dataset
print("Dataset Information:")
df.info()

# Preview the dataset
df.head()
```

```python
# Gender data
df["Gender"].tail()
```

```python
# Group the data by Gender
gender_grouped = df.groupby('Gender')['TotalClaims'].mean()

# Display the mean of TotalClaims for each Province
print(gender_grouped)
```

```python
# Ensure that the Gender column has no missing values
df['Gender'] = df['Gender'].fillna('Unknown')

# Group A: TotalClaims for males
male_claims = df[df['Gender'] == 'Male']['TotalClaims'].dropna()

# Group B: TotalClaims for females
female_claims = df[df['Gender'] == 'Female']['TotalClaims'].dropna()

print(f"Number of Male Claims: {len(male_claims)}, Number of Female Claims: {len(female_claims)}")


# Check sizes of both groups
print(f"Male Claims Size: {len(male_claims)}")
print(f"Female Claims Size: {len(female_claims)}")

# Downsample the larger group to match the smaller group's size if needed
min_size = min(len(male_claims), len(female_claims))
male_sampled = male_claims.sample(n=min_size, random_state=42)
female_sampled = female_claims.sample(n=min_size, random_state=42)

from scipy.stats import ttest_ind

# Perform t-test
t_stat, p_value = ttest_ind(male_sampled, female_sampled, equal_var=False)

# Print t-statistic and p-value
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Determine if we reject the null hypothesis
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant risk difference between men and women.")
else:
    print("Fail to reject the null hypothesis: There is no significant risk difference between men and women.")
```
